```python
import pandas as pd
from transformers import pipeline
```

```python

# Load the CSV file
url = "https://raw.githubusercontent.com/jogfx/MLops-exam/main/job_listings_cleaned.csv"
df = pd.read_csv(url)
```

```python
# Load categories from categories.txt
categories_file = "categories.txt"

categories = []

with open(categories_file, "r") as file:
    for line in file:
        if "/jobsoegning/" in line:
            category = line.strip().split("/jobsoegning/")[-1]
            categories.append(category)
```

```python

# Define the LLM pipeline for text classification
classifier = pipeline("zero-shot-classification")
```

```python

# Function to categorize a job post
def categorize_job_post(job_post):
    # Classify job post
    result = classifier(job_post, categories)
    
    # Get the top predicted category
    top_category = result["labels"][0]
    
    return top_category
```
